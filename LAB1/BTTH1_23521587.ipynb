```python
import pandas as pd
import numpy as np
```

```python
# load dữ liệu từ file ratings, movies
ratings = pd.read_csv('movies/ratings.csv')
movies = pd.read_csv('movies/movies.csv')
```

```python
# xem 1 cột trên file ratings
ratings['userId']
```

```python
# xem nhiều cột trên file ratings
ratings[['userId', 'movieId', 'rating']]
```

```python
# chọn ra các lượt rating > 3.5
ratings[ratings['rating'] > 3.5]
```

```python
# chọn ra các lượt rating cho bộ phim có id = 100
ratings[ratings['movieId'] == 100]
```

```python
# chọn ra các lượt rating của user có id là 350
ratings[ratings['userId'] == 350]
```

```python
# Thêm một cột "birthday" vào dataframe ratings với giá trị ngẫu nhiên cho từng hàng là 1960 đến 2010
ratings['birthday'] = np.random.randint(1960, 2010, size=len(ratings))
ratings['birthday']

```

```python
# thêm côt "is_old" vào dataframe ratings với giá trị True nếu năm <1990 và false nếu ngược lại
ratings['is_old'] = ratings['birthday'] < 1990
ratings

```

```python
# Xoá hai cột vừa tạo ở trên
ratings = ratings.drop(columns=['birthday', 'is_old'])
ratings
```

```python
# đếm số lượng rating cho từng giá trị rating 0.5 -> 5
ratings['rating'].value_counts().sort_index()
```

```python
# Vẽ biểu đồ cột thể hiện số lượng cho từng giá trị ratings trên
import matplotlib.pyplot as plt
ratings['rating'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Number of Ratings for Each Rating Value')
plt.show()
```

```python
# sắp xếp danh sách ratings theo chiều tăng dần, giảm dần theo giá trị rating
ratings.sort_values(by='rating', ascending=True)  # tăng dần
ratings.sort_values(by='rating', ascending=False) # giảm dần
```

```python
# đếm xem có bao nhiều người tham gia đánh giá
ratings['userId'].nunique()
```

```python
# Đếm xem có bao nhiêu bộ phim được đánh giá
ratings['movieId'].nunique()
```

```python
# Tính xem trung bình mỗi người rating bao nhiêu bộ phim
ratings['movieId'].count() / ratings['userId'].nunique()
```

```python
# Tính xem mỗi bộ phim có bao nhiêu lượt rating
ratings['userId'].count() / ratings['movieId'].nunique()
```

```python
# groupby bảng ratings theo "movieId" và tính rating trung bình cho từng bộ phim
ratings.groupby('movieId')['rating'].mean()
```

```python
# Tìm tên 5 bộ phim có rating trung bình thấp nhất
lowrate_movie = pd.merge(
  ratings.groupby('movieId')['rating'].mean().nsmallest(5), 
  movies, 
  on='movieId'
)
lowrate_movie
```

```python
# Tìm tên 5 bộ phim có rating trung bình cao nhất
highrate_movie = pd.merge(
  ratings.groupby('movieId')['rating'].mean().nlargest(5), 
  movies, 
  on='movieId'
)
highrate_movie
```

```python
# merge bảng ratings, và movies theo cột "movieID"
ratings_and_movies_df = pd.merge(ratings, movies, on='movieId')
ratings_and_movies_df
```

```python
# Chuyển bảng ratings sang dataframe với các cột là "userId" và các hàng là "movieId" (*)
ratings_df = ratings.pivot(index='movieId', columns='userId')['rating']
ratings_df
```

```python
# chuyển (*) sang numpy.ndarray (như ma trân rating) (**)
ratings_matrix = ratings_df.to_numpy(na_value=0)
ratings_matrix
```

```python
# Nhân ma ma trận trên với 2
ratings_matrix * 2
```

```python
# Cộng ma ma trận trên với 3
ratings_matrix + 3
```

```python
# lấy ra dòng đầu tiên trong (*) và chuyển sang numpy.ndarray (như vector cho movieId = 1)
first_row = ratings_df.iloc[0].to_numpy(na_value=0)
first_row
```

```python
# lấy ra dòng thứ 2 trong (*) và chuyển sang numpy.ndarray (như vector cho movieId = 2)
second_row = ratings_df.iloc[1].to_numpy(na_value=0)
second_row
```

```python
# tính tích vô hướng của hai vector trên
np.dot(first_row, second_row)
```

```python
# tính giá trị cosine giữa hai vector trên
cosine_similarity = np.dot(first_row, second_row) / (np.linalg.norm(first_row) * np.linalg.norm(second_row))
cosine_similarity
```

```python
# tích của vector movieId = 1 với ma trận (**)
first_row.dot(ratings_matrix.T)
```

# Book dataset

```python
# load dữ liệu từ file ratings, books
book_ratings = pd.read_csv('books/ratings.csv')
books = pd.read_csv('books/books.csv')
```

```python
# xem thông tin file ratings
book_ratings.info()
```

```python
book_ratings.columns
```

```python
# xem 1 cột "User-ID" trên file ratings
book_ratings['User-ID']
```

```python
# xem cột "ISBN" và "Book-Rating" trên file ratings
book_ratings[['ISBN', 'Book-Rating']]
```

```python
# chọn ra các lượt rating > 5
book_ratings[book_ratings['Book-Rating'] > 5]
```

```python
# chọn ra các lượt rating cho cuốn sách có id = 440414121
book_ratings[book_ratings['ISBN'] == '440414121']
```

```python
# chọn ra các lượt rating của user có id là 276848
book_ratings[book_ratings['User-ID'] == 276848]
```

```python
# thêm côt "is_new" vào dataframe ratings với giá trị True nếu Year-Of-Publication > 2000 và false nếu ngược lại
ratings_books_df = pd.merge(book_ratings, books[['ISBN', 'Year-Of-Publication']], 
                            on='ISBN', how='left')

ratings_books_df['Year-Of-Publication'] = pd.to_numeric(
    ratings_books_df['Year-Of-Publication'], errors='coerce'
)

book_ratings['is_new'] = ratings_books_df['Year-Of-Publication'] > 2000
book_ratings
```

```python
# Xoá cột vừa tạo ở trên
book_ratings = book_ratings.drop(columns=['is_new'])
```

```python
# đếm số lượng rating cho từng giá trị rating 0.5 -> 10
book_ratings['Book-Rating'].value_counts().sort_index()
```

```python
# Vẽ biểu đồ cột thể hiện số lượng cho từng giá trị ratings trên
book_ratings['Book-Rating'].value_counts().sort_index().plot(kind='bar')
plt.xlabel('Book-Rating')
plt.ylabel('Count')
plt.title('Number of Ratings for Each Book-Rating Value')
plt.show()
```

```python
# sắp xếp danh sách ratings theo chiều tăng dần, giảm dần theo giá trị rating
book_ratings.sort_values(by='Book-Rating', ascending=True)  # tăng dần
book_ratings.sort_values(by='Book-Rating', ascending=False) # giảm dần
```

```python
# đếm xem có bao nhiều người tham gia đánh giá
print(book_ratings['User-ID'].nunique())
# Đếm xem có bao nhiêu cuốn sách được đánh giá
print(book_ratings['ISBN'].nunique())
# Tính xem trung bình mỗi người rating bao nhiêu cuốn sách
print(book_ratings['ISBN'].count() / book_ratings['User-ID'].nunique())
# Tính xem mỗi bộ phim có bao nhiêu lượt rating
print(book_ratings['User-ID'].count() / book_ratings['ISBN'].nunique())
```

```python
# groupby bảng ratings theo "ISBN" và tính rating trung bình cho từng cuốn sách
book_ratings.groupby('ISBN')['Book-Rating'].mean()
```

```python
# Tìm tên 10 cuốn sách có rating trung bình thấp nhất
book_ratings.groupby('ISBN')['Book-Rating'].mean().nsmallest(10),
```

```python
# Tìm tên 10 cuốn sách có rating trung bình cao nhất
book_ratings.groupby('ISBN')['Book-Rating'].mean().nlargest(10),
```

```python
# merge bảng ratings, và books theo cột "ISBN"
book_ratings_df = pd.merge(book_ratings, books, on='ISBN')
book_ratings_df
```

```python
# Chuyển bảng ratings sang dataframe với các cột là "User-ID" và các hàng là "ISBN" (*)
# vì dữ liệu quá lớn nên em chỉ lấy 1/100 dữ liệu ban đầu
new_shape = int(book_ratings.shape[0]/100)
book_ratings_small = book_ratings.iloc[:new_shape,:]

book_ratings_small_df = (
    book_ratings_small
    .pivot(index="ISBN", columns="User-ID", values="Book-Rating")
    .reindex(book_ratings_small["ISBN"].unique())   # giữ đúng thứ tự ISBN gốc
)

book_ratings_small_df
```

```python
# chuyển (*) sang numpy.ndarray (như ma trân rating) (**)
book_ratings_small_matrix = book_ratings_small_df.to_numpy(na_value=0)
book_ratings_small_matrix
```

```python
# Nhân ma ma trận trên với 2.5
print(book_ratings_small_matrix * 2.5)
# Cộng ma ma trận trên với 1
print(book_ratings_small_matrix + 1)
```

```python
# lấy ra dòng đầu tiên trong (*) và chuyển sang numpy.ndarray (như vector cho ISBN = 034545104X)
first_row = book_ratings_small_df.iloc[0].to_numpy(na_value=0)
# lấy ra dòng thứ 2 trong (*) và chuyển sang numpy.ndarray (như vector cho ISBN = 155061224)
second_row = book_ratings_small_df.iloc[1].to_numpy(na_value=0)
# tính tích vô hướng của hai vector trên
np.dot(first_row, second_row)
```

```python
# tính giá trị cosine giữa hai vector trên
cosine_similarity = np.dot(first_row, second_row) / (np.linalg.norm(first_row) * np.linalg.norm(second_row))
```

```python
# tích của vector ISBN = 155061224 với ma trận (**)
second_row.dot(book_ratings_small_matrix.T)
```
