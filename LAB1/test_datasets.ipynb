```python
import pandas as pd

# Giả sử DataFrame của bạn tên là df
# Và cột nhóm là 'main_section', cột muốn kiểm tra NaN là 'content'

nan_counts = df.groupby('main_section')['content'].apply(lambda x: x.isna().sum())
print(nan_counts)
```

```python
nan_summary = df.groupby('title').apply(lambda x: x.isna().sum())
print(nan_summary)
```
